## 1.- Factorial
El **factorial** de un entero positivo n, **el factorial de n** o **n factorial** se define como el producto de todos los números enteros positivos desde 1 hasta n.
\begin{equation}
n! = 1  \times  2  \times  3  \times \cdots    \times  n 
\end{equation}

*Defina la función factorial*

```python
# Definimos la función factorial
def factorial(n): # Definimos la función factorial, cuya entrada en un número natural
    f,i=1,1       # Hacemos la asignación de las variable f y i ambas igual a 1
                  # f será el factorial del numero deseado e i es una variable para generar f  
    if n==0 or n ==1: # Definimos el factorial de 0 y 1 que es 1 en ambos casos
        return f      # Regresa el f que es el factorial
    else:             # Se calcula y se construye el factorial de un número diferente de 1 y 0
        while i <= n: # Mientras i es menor que el número del cual necesitamos su factorial
            f=f*i     # Se calcula el factorial de i
            i=i+1     # Se itera i en una unidad
        return f      # Regresa el factorial de n
```

```python
factorial(3) # Se comprueba la función factorial
```

```python
# Definimos el factorial para una lista
def facLista(A):        # Definimos la funcíon facList, cuya entrada es una lista
    D =[]               # Definimos C,D dos listas vacías
    for a in A:         # Para los elementos de las lista se hace lo siguiente
        d=factorial(a)  # Se calcula el factorial de la entrada a-esima de lista A en d
        D.append(d)     # d se apenda en la lista D.
    return D            # Regresa la lista con los factoriales de cada elemento de ella
```

```python
facLista([1,6,8,9,0,10]) # Se revisa la función facLista
```

## 2.-Suma de factoriales
Escriba un programa  que calcule y visualice 
\begin{equation}
1! + 2! + 3! + \cdots + (n-1)! + n!
\end{equation}
con $n \in \mathbb{N}$

```python
# Definimos la suma de n factoriales
def SF(n):             # Definimos la funcíon SF, cuya entrada es un número natural
    sm,i=0,1           # Asignación de las variable sm y i igual a 0 y 1 respectivamente
    while i<=n:        # Mientras i es menor que el número del cual necesitamos la suma de los primeros 
                       # n factoriales        
        f=factorial(i) # Calcula el factorial de i, llamando la función factorial
        sm=sm+f        # Se suma el valor de sm con f lo cual da el nuevo valor de sm
        i=i+1          # Se itera i en una unidad
    return sm          # Regresa la suma de los primero n factoriales
```

```python
SF(4) # Se revisa la función SF
```

## 3. Función seno
El **seno** de un ángulo $x$, en un triángulo rectángulo, se define como la razón entre el cateto opuesto al ángulo y la hipotenusa:
\begin{equation}
\sin x=\frac{a}{c} 
\end{equation}
<img src="ft.png">
Con $a$ el cateto opuesto y $c$ la hipotenusa. Pero el seno de un ángulo también se define como:

\begin{equation}
\sin x = \sum^{\infty}_{n=0} \; (-1)^n \; \dfrac{x^{2n+1}}{(2n+1)!}
\end{equation}

De tal forma que se puede aproximar hasta cierto grado *n*

```python
# Definimos la función seno
def funcionseno(x,n): # Definimos la función funcionseno cuya entradas son 
                      # el ángulo x y n el grado de la aproximación 
    s,i=0,0           # Hacemos la asignación de las variable s y i ambas igual a 0
    while i<=n:       # Mientras i es menor que el grado n al cual queremos trucar la serie
        t=(((-1)**i)*x**(2*i+1))/(factorial(2*i+1)) # La regla de operación para cada termino
        s=s+t         # Se suma el valor de s con t lo cual da el nuevo valor de s
        i=i+1         # Se itera i en una unidad
    return s          # Regresa el seno de x
```

## 4.- Serie  Geometrica
Para $r\neq 1$, la suma de los primeros n términos de una serie geométrica es:

\begin{equation}
   a + ar + a r^2 + a r^3 + \cdots + a r^{n-1} =
   \sum_{k=0}^{n-1} ar^k =
   a \, \frac{1-r^{n}}{1-r}
\end{equation}
donde $a$ es el primer término de la serie y r la razón común.

```python
### Definimos la serie geometrica
def seriegeo(n,r,a): # Definamos seriegeo cuyas entradas son el grado "n", el número "r"
                     # y el factor "a"
    i,suma=0,0       # Hacemos la asignación de las variable i y sumo ambas igual a 0
    if r==1:         # Si r es igual a cero que haga lo siguiente
        print "Error :´( r no puede ser 1" # Imprime "Error :´( r no puede ser 1"
    else:            # De lo contrario 
        while i<=(n-1):                # Mientras i sea menor que n-1 que haga lo siguiente
            t= a*(r**i)                # Que t sea el sumando 
            suma=suma+t                # A suma sumale suma más t y que esto sea suma
            i=i+1                      # Iteramos i en una unidad
        return suma, a*((1-(r**n))/(1-r)) # Regresa la serie y su comprobación
    
```

```python
seriegeo(10,2,4)
```

## 3.- Promedio, Varianza y Desviación estandar.
Realice un programa que calcule lo siguiente, de una lista:
- El promedio  $$\overline{X} = \frac{1}{n} \sum_{i=1}^{n} a_i = \frac{a_1+a_2+\cdots+a_n}{n} $$
- Varianza $$ s_n^2 = \frac 1n \sum_{i=1}^n \left(X_i - \overline{X} \right)^ 2 = \left(\frac{1}{n} \sum_{i=1}^{n}X_i^2\right) - \overline{X}^2$$
- Desviación estandar $\sigma = \sqrt{s_n^2}$

```python
def promedio(A):
    s,i=0,0
    n=len(A)
    while i<n:
        a=A[i]
        s=s+a
        i=i+1
    s=s*1.0
    p=s/n
    return p
```

```python
promedio([1,2,3,4])
```

```python
def varianza(B):
    X=promedio(B)
    i,S=0,0
    n=len(B)
    while i<n:
        b=B[i]
        p=(b-X)**2
        S=S+p
        i=i+1
    S=S*1.0
    Q=S/n
    print Q  
```

## 4.-Torres de Hanoi
Este juego tiene sus orígenes en la cultura oriental y es una leyenda sobre el Templo de Brahma. El juego, en su forma más tradicional, consiste en tres varillas verticales. En una de las varillas se apila un número indeterminado de discos. Los discos se apilan sobre una varilla en tamaño decreciente. No hay dos discos iguales, y todos ellos están apilados de mayor a menor radio en una de las varillas, quedando las otras dos varillas vacantes. El juego consiste en pasar todos los discos de la varilla ocupada (es decir la que posee la torre) a una de las otras varillas vacantes. Para realizar este objetivo, es necesario seguir tres simples reglas:

- Sólo se puede mover un disco cada vez.
- Un disco de mayor tamaño no puede descansar sobre uno más pequeño que él mismo.
- Sólo puedes desplazar el disco que se encuentre arriba en cada varilla.

<img src="TowerHanoi.png">

```python
def hanoi(I,C,F,n):
    if n == 1:
        print "Mueve el disco %g del poste %s al poste %s" %(n,I,F)
    else: 
        hanoi(I,F,C,n-1)
        print "Mueve el disco %g del poste %s al poste %s" %(n,I,F)
        hanoi(C,I,F,n-1)   
```

```python
hanoi("A","B","C",4)
```
